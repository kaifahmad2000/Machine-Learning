**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

Answer: Anamoly detection is the identification of observations, events or data points that deviate from what is usual, standard or expected, making them inconsistent with the rest of a data set.

Any data point which is placed at a distance from all normal data points is an anomaly. Hence anomalies are also called outliers.

Anomaly detection is also called as deviation detection because anomalous objects have attribute values that are different from all normal data objects.

Anomalous data can signal critical incidents happening under the hood, such as an infrastructure failure, a breaking change from an upstream source or security threats. Anomalies can also highlight opportunities for architectural optimization or improving marketing strategies.

**Types of Anomalies:**
1. Point anomalies: A sole instance of data is abnormal if it is too distant from the remainder. It is the most common type of anomaly.
Example: In credit card transaction: a sudden purchase of 1,00,000/- rupees when usual spend is only 2000 - 5000 rupees only.

2. Contextual anomalies: These are data points that deviate from the norm within a specific context. These anomalies are not necessarily outliers when considered in isolation but become anomalous when viewed within their specific context.
For example, consider home energy usage. If there is a sudden increase in energy consumption at midday when no family members are typically home, the anomaly would be contextual. This data point might not be an outlier when compared to energy usage in the morning or evening (when people are usually home), but it is anomalous in the context of the time of day it occurs.
To tackle such anomalies, Isolation Forest Algorithm are used.

3. Collective anomalies: It involve a set of data instances that together deviate from the norm, even though individual instances may appear normal. Example: a network traffic data set that shows a sudden surge in traffic from multiple IP addresses at the same time.

**Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

Answer:
1) Isolation Forest:
- works on the principle that anomalies are easier to isolate than normal points.
- randomly selects a feature and splits data; anomalies need fewer splits to isolate.
- uses an ensemble of random trees.
use cases: high dimensional datasets, fraud detection in finance, and detecting anomalies in large-scale datasets(server logs, credit
card data.)

2. DBSCAN(Density Based Spatial Clustering of Applications with Noise):
- a clustering algorithm that groups points into dense regions.
- points that do not belong to any cluster are treated as anomalies.
- based on two parameters: epsilon (neighborhood size) and minimum points(minimum samples to form cluster).
use cases: spatial or geographical data like: GPS tracking, earthquake detection, identifying noise in sensor data, and works well when clusters have irregular shapes.

3. Local Outlier Factor:
- measures the local density deviation
 of a point compared to its neighbors.
- a point is anomalous if its density is significantly lower than that of its neighbors.
- more sesnsitive to local patterns than global ones.
use cases: detecting anomalies in datasets with varying density regions, cybersecurity, and healthcare.

**Question 3: What are the key components of a Time Series? Explain each with one example.**

Answer: Time series analysis is a way of studying the characteristics of the response variable concerning time as the independent variable. To estimate the target variable in predicting or forecasting, use the time variable as the reference point. Time series analysis represents a series of time based orders, it would be Years, Months, Weeks, Days, Hours, Minutes, and Seconds. It is an observation from the sequence of discrete time of successive intervals.

**Components of a Time Series**
1. Level: It is the base value of the series on which other components are added. Think it as the mean or central tendency of the series at a given point of time.
Example: If a sales data moves around a certain unit for most of the time, then it is the level.

2. Trend: It shows the general direction of the data, and whether it is increasing, decreasing or remaining stationary over an extended period of time. Trends indicate the long term movement in the data and can reveal overall growth or decline.
In simple words, we can say that a long term upward or downward movement in the data over time. Example:
The steady increase in global average temperature over the last century due to climate change.

3. Seasonality: It refers to predictable patterns that recur regularly, like yearly retail spikes during the holiday season. Seasonal components exhibit fluctuations fixed in timing, direction and magnitude. Example: Ice cream sales increase every summer and decline in winter.

4. Cycles: Cycles demonstrate fluctuations that do not have a fixed period, such as economic expansions and recessions. These long term patterns last longer than a year and do not have consistent amplitudes or durations. Example: Business cycles that oscillate between growth and decline.

5. Noise: Noise encompasses the residual variability in the data that the other components cannot explain. Noise includes unpredictable, erratic deviations after accounting for trends, seasonality and cycles. Example: A sudden drop in airline travel during covid-19 pandemic.

**Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?**

Answer: A stationary series is one whose statistical properties such as mean, variance, covariance, and standard deviation do not vary with time. In other words, stationarity in time series also means series without a Trend or Seasonal components. There are three main types of stationary series:
- Strict Stationary: Satisfies the mathematical definition of a stationary process. Mean, variance & covariance are not a function of time.
- Seasonal Stationary: Series exhibiting seasonality.
- Trend Stationary: Series exhibiting trend.

**Testing**

The most basic methods for stationarity check rely on plotting the data and visually checking for trend and seasonal components. Trying to determine whether a stationary process generated a time seires just by looking is technically not possible, what we will do is to check some basic properties of non-stationary data.
To check if a series is non-stationary we can perform any of the two statistical tests:
1. Augmented Dickey Fuller(ADF).
2. Kwiatkowski-Phillips-Schmidt-Shin(KPSS).

ADF test is used to see if a time series is stationary. Here is how we perform the test:
- The Hypothesis: The test has a null hypothesis that the data has a unit root, which means it's not stationary. The alternative hypothesis is that the data is stationary or trend-stationary.
- The Output: The ADF test gives you a test statistic, a p-value, and critical values at different significance levels (like 1% or 5%).
- Interpreting the p-value: A low p-value (less than your chosen significance level) means you reject the null hypothesis. In this case, that means the data is likely stationary.
- Interpreting the test statistic: The test statistic is negative. The more negative it is, the stronger the evidence against the null hypothesis (i.e., stronger evidence for stationarity)

The KPSS test is a type of Unit root test that tests for the stationarity of a given series around a deterministic trend. In other words, the test is somewhat similar in spirit to the ADF test. A common misconception, however, is that it can be used interchangeably with the ADF test. This can lead to misinterpretations about stationarity, which can easily go undetected, causing more problems down the line.

Ideally we have to perform both ADF and KPSS tests and get the four cases:
- Both test conclude that the given series is stationary.
- Both test conclude that the given test is non-stationary.
- ADF concludes non-stationary and KPSS concludes stationary. The series is trend stationary, to make the series strictly stationary, we need to remove the trend in this case. Then we check the detrended series for stationarity.
- ADF concludes stationary, and KPSS concludes non-stationary. The series is difference stationary. Differencing is used to make series stationary. Then we check the differnced series for stationarity.

**Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.**

Answer:
1. **AR**: Auto regression model is a machine learning technique most commonly used for time series analysis and forecasting that uses one or more values from previous time steps in a time series to create a regression.
It is a simple but powerful technique for time series analysis that provides highly interpretable and effective predictions if your data contains correlations across the time steps. The correlation across time steps is called autocorrelation because it is a measure of how much a value correlates with itself. A purely linear process will autocorrelate perfectly with itself across the time series, making it possible to predict the next value exactly from previous values using an autoregressive process.
An autoregressive model is when we regress a value from a time series on previous values from that same time series. For example, yt regressed on yt-1 uses the previous value of y, called a lagged value, to predict the current value of y. In this simple regression model, the dependent variable in the previous time period has become the predictor. The errors represent all the usual assumptions about errors in a simple linear regression model. We often look at the order of an autoregression as the number of preceding values in the series used to predict the value now. So, yt regressed on yt-1 is a first-order autoregression, which is written as AR(1).
- Equation:

    yt = c+ϕ1​yt−1​+ϕ2​yt−2​+⋯+ϕp​yt−p​+ϵt

  where, yt = value at current time stamp,
  ϕ = coefficient,
  c = cosntant,
  ϵt = error term

2. **Moving Average**: Moving Average model in time series is not the same as the moving average smoothing technique.
- In MA model, the current observation is expressed as a linear combination of past error terms.
- It is a stochastic model, used for time series forecasting.

  yt=μ+ϵt+θ1ϵt-1+θ2ϵt-2+⋯+θqϵt-q

  where, yt = value of the time series at time t,
  μ = mean of the series,
  ϵt = white noise,
  θi = coefficients,
  q = order of the MA model

- Unlike AR models that depends on past values, MA models depends on past errors.
- If there was a sudden shock yesterday(like: extreme demand, strike or certain economic news), it effect continues for a few future periods but then fades away.

3. **ARIMA**: Autoregressive Integrated Moving Average is a technique for time series analysis and for forecasting possible future values of a time series.
ARIMA is one of the most widely used approaches to time series forecasting and it can be used in two different ways depending on the type of time series data that you're working with. In the first case, we have create a Non-seasonal ARIMA model that doesn't require accounting for seasonality in your time series data. We predict the future simply based on patterns in the past data. In the second case, we account for seasonality which is regular cycles that affect the time series. These cycles can be daily, weekly, or monthly and they help define patterns in the past data of the time series that can be used to forecast future values. Like much of data science, the foundation of forecasting is having good time series data with which to train your models. A time series is an ordered sequence of measurements of a variable at equally spaced time intervals. It's important to remember that not all data sets that have a time element to it is actually time series data because of this equally spaced time interval requirement.

4. **SARIMA**: Seasonal Autoregressive Integrated Moving Average adds seasonality. The parameters for seasonality are represented with capital letters and the parameter m indicates the duration of the season.
SARIMA model are denoted as (p,d,q) * (P,D,Q,m) and requires a large quantity of historical data.

5. **SARIMAX**: SARIMA with exogenous variable known as SARIMAX. More complex time series data includes variables in addition to time. SARIMAX models incorporate external variables to generate more nuanced forecasts. It provides advantage over that of SARIMA:
- The implemetation is fairly easy to set up and use.
- The structure is same for other forecasting models.
- Even though SARIMAX is an old methodology, it still gets super good results.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import LocalOutlierFactor
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')
```

```python
# Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
# and decompose it into trend, seasonality, and residual components.

# data preprocessing
df = pd.read_csv('AirPassengers.csv')
pass_data = df[['Month', '#Passengers']]
pass_data['Month'] = pd.to_datetime(pass_data['Month'])
print(pass_data.describe())

# plotting the original series
plt.style.use('ggplot')
plt.figure(figsize = (18,8))
plt.grid(True)
plt.xlabel('Month', fontsize = 20)
plt.xticks(fontsize = 15)
plt.ylabel('#Passengers', fontsize = 20)
plt.yticks(fontsize = 20)
plt.plot(pass_data['Month'], pass_data['#Passengers'], linewidth = 3, color = 'blue')
plt.title('Number of Passengers travelled')
plt.show()

#components of time series
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(pass_data['#Passengers'], period = 12)

# plotting of time series
fig = plt.figure(figsize = (18,8))
fig = result.plot()
fig.set_size_inches(17,10)
plt.show()
```

```python
# Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
# detect anomalies. Visualize the anomalies on a 2D scatter plot.

#data ingesion/ preprocessing
df_1 = pd.read_csv('NYC_taxi_fare_data.csv')
#  handling missing values
df_1.isnull().sum()
df_1.dropna(inplace = True)
# convert object to integer and date time series
df_1['tpep_pickup_datetime'] = pd.to_datetime(df_1['tpep_pickup_datetime'])
df_1['tpep_dropoff_datetime'] = pd.to_datetime(df_1['tpep_dropoff_datetime'])
df_1 = pd.get_dummies(df_1, columns = ['store_and_fwd_flag'], drop_first = True)

# feature extraction from time series data
df_1['pickup_hour'] = df_1['tpep_pickup_datetime'].dt.hour
df_1['pickup_dayofweek'] = df_1['tpep_pickup_datetime'].dt.dayofweek
df_1['trip_duration_minutes'] = (df_1['tpep_dropoff_datetime'] - df_1['tpep_pickup_datetime']).dt.total_seconds() / 60
df_1 = df_1.drop(columns = ['tpep_pickup_datetime', 'tpep_dropoff_datetime'])


# model training
from sklearn.ensemble import IsolationForest
model = IsolationForest(random_state = 1)
model.fit(df_1)

df_1['anomaly'] = model.predict(df_1)
num_anomalies = df_1[df_1['anomaly'] == -1].shape[0]
print(f"The number of anomalies: {num_anomalies}")

# plotting
fig, axes = plt.subplots(1, 2, figsize = (18,8))
axes[0].scatter(df_1['trip_distance'][df_1['anomaly'] == 1], df_1['fare_amount'][df_1['anomaly'] == 1],color = 'blue', alpha = 0.1)
axes[0].set_title('Normal Data Points', fontsize = 15)
axes[0].set_xlabel('trip distance')
axes[0].set_ylabel('fare amount')
axes[1].scatter(df_1['trip_distance'][df_1['anomaly'] == -1], df_1['fare_amount'][df_1['anomaly'] == -1], color = 'red', alpha = 0.6)
axes[1].set_title('Outliers', fontsize = 15)
axes[1].set_xlabel('trip distance')
axes[1].set_ylabel('fare amount')

plt.tight_layout()
plt.show()
```

```python
# Question 8: Train a SARIMA model on the monthly airline passengers dataset.
# Forecast the next 12 months and visualize the results.

# data preprocessing
df_2 = pd.read_csv('AirPassengers.csv')
# convert object into date time and make it as index
df_2['Month'] = pd.to_datetime(df_2['Month'])
df_2.set_index('Month', inplace = True)
df_2

# specify SARIMA constraints

order = (1, 1, 1) # non-seasonal order
seasonal_order = (1, 1, 1, 12)

# model training and forecasting
sarima_model = sm.tsa.statespace.SARIMAX(df_2['#Passengers'], order = order, seasonal_order = seasonal_order)
results = sarima_model.fit()
results

forecast = results.get_forecast(steps = 12)
forecast_index = pd.date_range(df_2.index[-1] + pd.DateOffset(months = 1), periods = 12, freq = 'MS')
forecasted_values = forecast.predicted_mean.values
forecasted_values = [int(x) for x in forecasted_values]
forecasted_values

forecasted_df = pd.DataFrame({'#Passengers': forecasted_values}, index = forecast_index)
combined_df = pd.concat([df_2, forecasted_df])

#plotting
fig, axes = plt.subplots(1, 2, figsize = (18,8))
axes[0].set_title("Actual with Forecasted Values", fontsize = 25)
axes[0].plot(df_2.index, df_2,label = "Actual", color = 'blue')
axes[0].plot(forecasted_df.index, forecasted_df['#Passengers'], label = 'Forecasted', color = 'red')
axes[0].set_xlabel('Month', fontsize = 20)
axes[0].set_ylabel('#Passengers', fontsize = 20)
axes[0].legend()

axes[1].set_title("Forecasted Values", fontsize = 25)
axes[1].plot(forecasted_df.index, forecasted_df['#Passengers'], label = 'Forecasted', color = 'red')
axes[1].set_xlabel('Month', fontsize = 20)
axes[1].set_ylabel('#Passengers', fontsize = 20)

plt.tight_layout()
plt.show()
```

```python
# Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
# anomalies and visualize them using matplotlib.

# data preprocessing
from sklearn.datasets import make_moons
X, y = make_moons(n_samples = 300, noise = 0.3)

x_outliers = np.random.uniform(low = -3, high = 3, size = (20,2))
X = np.vstack((X, x_outliers))

# model training
loc = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)
y_pred = loc.fit_predict(X)

# seperating inliers and outliers
X_inliers = X[y_pred == 1]
X_outliers = X[y_pred == -1]

#plotting
plt.title("Local Outlier Factor on moons data")
plt.scatter(X_inliers[:,0], X_inliers[:,1], color = 'blue', s = 20, edgecolor = 'k', label = 'Normal Data Points')
plt.scatter(X_outliers[:,0], X_outliers[:,1], color = 'red', s = 20, edgecolor = 'k', label = 'Outliers')
plt.legend()
plt.show()
```
