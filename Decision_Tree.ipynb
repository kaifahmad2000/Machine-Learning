Question 1: What is a Decision Tree, and how does it work in the context of
classification?
Answer: A Decision Tree is a supervised machine learning algorithm used for discrete and continuous datasets.

Root Node: Starting point representing the whole dataset.

Branches: Lines connecting nodes showing the flow from one decision to another.

Internal Nodes: Points where decisions are made based on data features.

Leaf Nodes: End points of the tree where the final decision or prediction is made.

There are mainly two types of Decision Tree based on the target variable:
1. Classification Tree: It is used for predicting categorical outcomes like spam or ham. These trees split the data based on features to classify data into predefined categories.
2. Regression Trees: It is used to predict continuous outcomes like predicting house prices. Instead of assigning categories, it provides numerical predictions based on the input features.

Working of decision tree on classificaton:

1. It begins with a main question at the root node which is derived from the dataset's features.
2. From the root, the tree asks a series of yes/ no questions to split the data into subsets based on specific attributes.
3. Each question leads to different branches:
- If the answer is yes, the tree follows one path.
- If the answer is no, the tree follows another path.
4. This branching continues through further decisions helps in reducing the data down step-by-step.
5. The process ends when there are no more useful questions to ask leading to the leaf node where the final decision or prediction is made.

Let's take a simple example: Imagine we have to decide whether we want to play cricket based on the wheather conditions:
Root Node --> Outlook
if overcast --> always Yes
if sunny --> check for humidity
  - if humidity high --> No
  - if humidity normal --> Yes
if rainy --> check windy
  - if windy == True --> No
  - if windy == False --> Yes

Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures.How do they impact the splits in a Decision Tree?

Answer: In a decision tree, impurity refers to how mixed the data is at a particular node. A high impurity indicates the node a high variety of classes, while low impurity indicates the node is mostly made up of a single class. There are different ways to define impurity, generally we use gini impurity index and entropy.

Gini Impurity: It is linear measure and calculates the probability of misclassifying a randomly chosen element in a set. The range of gini index is [0, 0.5] where 0 indicates perfect purity and 0.5 indicates maximum impurity.

GI = 1 - summation of (pi)2 where i ranges from 1 to n.

Entropy: The expected volume of information, surprise or uncertainity associated with a randomly chosen variable's potential outcomes is characterized as the entropy of the variable in information theory.

Entropy = - summation of pilog2(pi) where i ranges from 1 to c.

Entropy is a logarithmic measure, measures the amount of uncertainity or randomness in a set. The range of entropy is [0, log2(c)] where c is the number of classes. The range becomes [0, 1] for binary classification.

At each node, the tree algorithm evaluates all possible splits across features and thresholds. For each split, it computes:
- parent impurity
- child impurity
- information gain

the split that maximizes impurity reduction is chosen.

Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each.

Answer: Pruning is a technique to prevent overfitting by reducing the size of the tree. There are two ways: Pre-Pruning and Post-Pruning.

1. Pre-Pruning(Early Stopping):
Sometimes the growth of the decision tree can be stopped before it gets too complex, this is called pre-pruning. It is important to prevent the overfitting of the training data, which results in a poor performance when exposed to new data.Some common pre-pruning techniques include:

- Maximum Depth: It limits the maximum level of depth in a decision tree.
- Minimum Samples per Leaf: Set a minimum threshold for the number of samples in each leaf node.
- Minimum Samples per Split: Specify the minimal number of samples needed to break up a node.
- Maximum Features: Restrict the quantity of features considered for splitting.

2. Post-Pruning(Reducing nodes):
After the tree is fully grown, post-pruning involves removing branches or nodes to improve the model's ability to generalize.Some common post-pruning techniques include:
- Cost-Complexity Pruning (CCP): This method assigns a price to each subtree primarily based on its accuracy and complexity, then selects the subtree with the lowest fee.
- Reduced Error Pruning: Removes branches that do not significantly affect the overall accuracy.
- Minimum Impurity Decrease: Prunes node­s if the decrease in impurity (Gini impurity or entropy) is beneath a certain threshold.
- Minimum Leaf Size: Removes leaf nodes with fewer samples than a specified threshold.

Question 4: What is Information Gain in Decision Trees, and why is it important for choosing the best split?

Answer: Information Gain tells us how useful a question (or feature) is for splitting data into groups. It measures how much the uncertainty decreases after the split. A good question will create clearer groups and the feature with the highest Information Gain is chosen to make the decision.

Suppose S is a set of instances and A is an attribute, Sv is the subset of S, v represents an individual value that the attribute A can take and value(A) is the set of all possible values of A then
IG(S,A) = Entropy(S) - summation of Sv/S
* Entropy(Sv)

IG is important because DT split nodes based on which feature gives the highest Information Gain. A high IG means:
- the feature reduces impurity the most
- leads to purer child nodes
- improves classificatoin accuracy.
in simple words, the split with the highest IG is chosen because it creates the most informative seperation of classes.

Question 5: What are some common real-world applications of Decision Trees, and what are their main advantages and limitations?

Answer: Some real world application of decision tree are:

1. Banking and Finance: Predict whether a loan application is default based on credit score and detect fraud based on unsual transactions.

2. Healthcare: Classify a patient as high or low risk based on symptoms and test results and recommend personalized treatments.

3. Marketing and Sales: Customer segregation to identify high value customers, predict which customer may leave(churn prediction) and campaign targetting.

4. Manufacturing and Operations: Maintain quality control by detecting defective products and predictive maintenance by predicting machine failures.

5. Technology: Classify emails as spam or ham and suggest product or content based on the search preferences.

- Advantages:
1. Easy to Understand & Interpret: Can be visualized like a flowchart.

2. Handles Both Numerical & Categorical Data.

3. No Need for Feature Scaling: Unlike SVM or Logistic Regression.

4. Captures Non-linear Relationships.

5. Fast Predictions: Once trained, classification is quick.

- Disadvantages:
1. Overfitting: Trees can become too complex and memorize training data.

2. Unstable: Small changes in data may lead to very different trees.

3. Biased Toward Features with More Levels e.g., splitting on an ID-like feature.

4. Less Accurate Alone: Often outperformed by ensemble methods like Random Forests or Gradient Boosting.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris, fetch_california_housing

import warnings
warnings.filterwarnings('ignore')
```

```python
# Question 6: Write a Python program to:
#● Load the Iris Dataset
#● Train a Decision Tree Classifier using the Gini criterion
#● Print the model’s accuracy and feature importances'''
# data ingestion
data = load_iris()
x = data.data
y = data.target
# train-test-split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)

#model training
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'gini', random_state = 1)
clf.fit(x_train, y_train)
from sklearn import tree
tree.plot_tree(clf, filled = True)

#Prediction
y_pred = clf.predict(x_test)

#accuracy
from sklearn.metrics import accuracy_score
print(f"The accuracy score of the model is {accuracy_score(y_test, y_pred)}")

#feature importance
print("Feature Importance:")
for feature, importance in zip(data.feature_names, clf.feature_importances_):
    print(f"{feature}: {importance}")

```

```python
#Question 7: Write a Python program to:
#● Load the Iris Dataset
#● Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to
#a fully-grown tree.
# Data preprocessing
iris = load_iris()
X = iris.data
Y = iris.target

#Train-test-split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 1)

# model training
pruned_tree = DecisionTreeClassifier(max_depth = 3, random_state = 1)
pruned_tree.fit(X_train, Y_train)
tree.plot_tree(pruned_tree, filled = True)

#prediction
Y_pred = pruned_tree.predict(X_test)

print(f"Accuracy of the fully grown tree: {accuracy_score(y_test, y_pred)}")
print(f"Accuracy of pruned tree: {accuracy_score(Y_pred, Y_test)}")
```

```python
# Question 8: Write a Python program to:
#● Load the California Housing dataset from sklearn
#● Train a Decision Tree Regressor
#● Print the Mean Squared Error (MSE) and feature importances
# data ingestion
data = fetch_california_housing()
x = data.data
y = data.target

#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)

#model_training
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 1)
regressor.fit(x_train, y_train)

#prediction
y_predict = regressor.predict(x_test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predict)

print("The MSE of the model is: ", mse)

print("Feature Importance:")
for feature, importance in zip(data.feature_names, regressor.feature_importances_):
    print(f"{feature}: {importance}")
```

```python
#Question 9: Write a Python program to:
#● Load the Iris Dataset
#● Tune the Decision Tree’s max_depth and min_samples_split using
#GridSearchCV
#● Print the best parameters and the resulting model accuracy
#data ingestion
iris = load_iris()
x, y = iris.data, iris.target

# train-test-split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
dt = DecisionTreeClassifier(random_state = 1)

#GridSearch Cross Validation
param = {
    'max_depth': [2, 3, 4, 5, 6],
    'min_samples_split': [2, 3, 4, 5, 10]
}

g_cv = GridSearchCV(
    estimator = dt,
    param_grid = param,
    cv = 5,
    verbose = 2,
    scoring = 'accuracy'
)

g_cv.fit(x_train, y_train)

print(f"Best Parameters: {g_cv.best_params_}")
print(f"Best Score: {g_cv.best_score_}")
```
